```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(np.__version__)
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
dataset=dataset.drop("User ID",axis=1)
```

```python
dataset.columns
```

```python
independent=dataset[["Gender_Male", "Age", "EstimatedSalary"]]
independent
```

```python
dependent=dataset[["Purchased"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid= {'n_estimators':[10,50,100], 
             'criterion': ['entropy','gini'], 
             'max_features': ['auto','sqrt','log2'],
             'random_state':[0]}
grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
print("The re for the best Parameter{}:".format(grid.best_params_),re)
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
grid_predictions=grid.predict(X_test)


from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,grid_predictions)

from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_micro=f1_score(Y_test,grid_predictions,average='weighted')
print("The f1_micro value for best parameter{}:".format(grid.best_params_),f1_micro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
